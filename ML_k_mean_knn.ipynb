### Supervised Machine Learning with new labels using TSNE to reduce dimensionality

#### 1. convert the wine class from string to integer so it can perform k-neighborsclassifier

```python
# convert the wine class from string to integer so it can use in further k-neighborsclassifier
relabeled_wine_df_split = relabeled_wine_df2.replace("class_0", 0)
relabeled_wine_df_split = relabeled_wine_df_split.replace("class_1", 1)
relabeled_wine_df_split = relabeled_wine_df_split.replace("class_2", 2)
```

```python
relabeled_wine_df_split
```

#### 2. split the dataset to train and test sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
     relabeled_wine_df_split.iloc[:,[1,2]],relabeled_wine_df_split["wine classification"], random_state=22)

print(f'X_train.shape = {X_train.shape}')
print(f'X_test.shape  = {X_test.shape}')

print()
print(f'y_train.shape = {y_train.shape}')
print(f'y_test.shape  = {y_test.shape}')
```

#### 3. perform k-neighborsClassigier with train dataset to build the predictive model

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X=X_train, y=y_train)
```

#### 4. make prediction with the predicitive model and x test set, and compare with the expected test dataset

```python
import numpy as np

predicted = knn.predict(X=X_test)
expected = np.array(y_test)
print('First twenty predictions:')
print(f'predicted[:20] = {predicted[:20]}')
print(f' expected[:20] = {expected[:20]}')
```

```python
wrong = [ (pred, exp) 
          for (pred, exp) in zip(predicted, expected) 
          if pred != exp
        ]

print('Wrong predictions(predicted, expected):')
print(wrong)
print(f'Prediction accuracy score = {knn.score(X_test, y_test):.2%}')
```

#### The model gives 75.56 % hit rate

#### 5. create a confusion matrix and the heatmap of the matrix

```python
from sklearn.metrics import confusion_matrix

confusion = confusion_matrix(y_true=expected, y_pred=predicted)

print('Confusion matrix:')
print(confusion)
```

```python
import pandas as pd
import seaborn as sns

confusion_df = pd.DataFrame(confusion, index=range(3), columns=range(3))
axes = sns.heatmap(confusion_df, annot=True, cmap='nipy_spectral_r')
```

#### 6. Take a look at the classification report

```python
from sklearn.metrics import classification_report

print(classification_report(expected, predicted, target_names=wine.target_names))
```

## Further experiments
Assuming that the wine dataset is not a cheating data that has already been labeled. I try to apply different number of cluster to kmeans clustering method and create scatter plot to see the clustering

```python
def dimensionalityReduction(n = 2):
    pca = PCA(n_components=n, random_state=22)  # reduce to two components
    reduced_wine = pca.fit_transform(wine.data)
    return reduced_wine
```

```python
def kmeans(reduced_wine, k = 3):
    kmeans = KMeans(n_clusters=k, random_state=22) 
    kmeans.fit(reduced_wine.data)
    relabeled_wine_df = pd.DataFrame(reduced_wine, 
                               columns=["Component "+str(i) for i in range(1, reduced_wine.data.shape[1]+1)])
    relabeled_wine_df["wine classification"] = kmeans.labels_
    return relabeled_wine_df 
```

```python
wine_data = dimensionalityReduction()
dflist = []
for i in range(3,11):
    relabeled_wine_df = kmeans(wine_data, i)
    dflist.append(relabeled_wine_df)
```

```python
for i in dflist:
    axe = sns.FacetGrid(i)
    axe.map(sns.scatterplot, data = i, hue = 'wine classification', legend = 'brief', 
                          x = 'Component 1', y = 'Component 2')
```
